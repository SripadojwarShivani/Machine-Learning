```python
company=["amazon","google","microsoft","facebook"]
revenue=[90,136,85,26]
plt.bar(company,revenue)
plt.ylabel("revenue($B)")
plt.title("US Technology stocks")
plt.show()
```

```python
company=["amazon","google","microsoft","facebook"]
revenue=[90,136,85,26]
profit=[40,20,30,12]
plt.bar(company,revenue,label="revenue")
plt.bar(company,profit,label="profit")
plt.title("US Technology Stocks")
plt.legend()
plt.show()
```

```python
company=["amazon","google","microsoft","facebook"]
revenue=[90,136,85,26]
profit=[40,20,30,12]

xpos = np.arange(len(company))
plt.bar(xpos-0.2,revenue, width=0.4, label="revenue")
plt.bar(xpos+0.2,profit,width=0.4, label="profit")

plt.xticks(xpos,company)
plt.ylabel("revenue(Bln)")
plt.title("US Technology Stocks")
plt.legend()
plt.show
```

```python
#multiple charts
ecommerce=['amazon','snapdeal','flipkart','myntra']
Q1_profit=['70','45','40','38']
Q2_profit=['65','40','45','38']
Q3_profit=['73','42','50','30']
Q4_profit=['60','41','48','25']
```

```python
#using subplot
plt.figure(1,figsize=(10,10))

#first cell
plt.subplot(2,2,1)
plt.bar(ecommerce,Q1_profit)
plt.title("Quarter1 profit")

#second cell
plt.subplot(2,2,2)
plt.bar(ecommerce,Q2_profit)
plt.title("Quarter2 profit")

#Third cell
plt.subplot(2,2,3)
plt.bar(ecommerce,Q3_profit)
plt.title("Quarter3 profit")

#fourth cell
plt.subplot(2,2,4)
plt.bar(ecommerce,Q4_profit)
plt.title("Quarter4 profit")

#Adding title
plt.suptitle("profit on Quarterly Basis")

#Displaying the chart
plt.show()
```
